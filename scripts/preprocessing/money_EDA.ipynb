```python
import pandas as pd
file2 = 'preprocessed_cars.csv'
df = pd.read_csv(file2,delimiter=';')
```

```python
df
```

```python
df = df.rename(columns={'check': 'direction'})
```

```python
labels = ["자린고비", "가성비", "평범", "욜로", "플렉스"]

df_come = df[df['direction'] == 'come'].copy()
df_back = df[df['direction'] == 'back'].copy()

df_come.sort_values('adult_charge', inplace=True)
df_come['금전상황'] = pd.qcut(df_come['adult_charge'], q=5, labels=labels)

df_back.sort_values('adult_charge', inplace=True)
df_back['금전상황'] = pd.qcut(df_back['adult_charge'], q=5, labels=labels)


df_new = pd.concat([df_come, df_back])

df_new.sort_index(inplace=True)

df_new.head()
```

```python
def safe_qcut(x):
    unique_values = x.nunique()
    if unique_values >= 5:
        return pd.qcut(x, q=5, labels=labels, duplicates='drop')
    else:
        new_labels = labels[:unique_values]
        return pd.qcut(x, q=unique_values, labels=new_labels, duplicates='drop')

# Convert 'start_date' and 'end_date' to datetime
df['start_date'] = pd.to_datetime(df['start_date'])
df['end_date'] = pd.to_datetime(df['end_date'])

# Step 3: Define the labels
labels = ["자린고비", "가성비", "평범", "욜로", "플렉스"]

# Step 4: Apply safe_qcut() to 'price' within each group of 'start_date' and 'end_date'
df['금전상황'] = df.groupby(['start_date', 'end_date'])['price'].transform(safe_qcut)

# Display the first few rows of the dataframe
df
```
